```python
import pandas as pd
```

```python
file_apple='precios/AAPL.csv'
df_apple=pd.read_csv(file_apple,index_col='Date', usecols=['Date','Adj Close'])
df_apple
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_apple.plot(figsize=(8,6))
```

```python
file_amazon='precios/AMZN.csv'
df_amazon=pd.read_csv(file_amazon,index_col='Date', usecols=['Date','Adj Close'])
file_micro='precios/MSFT.csv'
df_micro=pd.read_csv(file_micro,index_col='Date', usecols=['Date','Adj Close'])
file_vidia='precios/NVDA.csv'
df_vidia=pd.read_csv(file_vidia,index_col='Date', usecols=['Date','Adj Close'])
```

```python
df_apple.plot(figsize=(8,6),title='Apple')
df_amazon.plot(figsize=(8,6),title='Amazon')
df_micro.plot(figsize=(8,6),title='Microsoft')
df_vidia.plot(figsize=(8,6),title='Nvidia')
```

```python
closes=pd.DataFrame(index=df_amazon.index,columns=['AAPL','AMZN','MSFT','NVDA'])
closes.index.name='Date'
closes['AAPL']=df_apple
closes['AMZN']=df_amazon
closes['MSFT']=df_micro
closes['NVDA']=df_vidia
closes
```

```python
from pandas_datareader import data
import datetime
```

```python
tickers=['AAPL','AMZN','MSFT','NVDA']
data_source='yahoo'
start_date=datetime.datetime(2011,1,1)
end_date=datetime.datetime(2016,12,31)
panel_data=data.DataReader(tickers,data_source,start_date,end_date)
```

```python
panel_data
```

```python
closes=panel_data.loc['Adj Close']
closes
```

```python
closes.plot(figsize=(8,6))
```

```python
pd.set_option('display.max_columns',6)
pd.set_option('display.max_rows',10)
pd.set_option('display.width',78)
pd.set_option('precision',3)
closes
```
