```python
%matplotlib inline
import matplotlib.pyplot as plt
from sympy import *
import pandas as pd
import numpy as np
```

```python
var('x,t')
plot(sin(x),(x,0,pi))
integrate(sin(x),(x,0,pi))
```

```python
n=1000
f=lambda x: np.sin(x)
a,b=0,np.pi
altura=1
x=np.random.uniform(a,b,n)
y=np.random.uniform(a,altura,n)
X=np.arange(a,b,.01)
fx=f(x)
plt.plot(x[y>fx],y[y>fx],"go")
plt.plot(x[y<fx],y[y<fx],"ro")
plt.plot(X,f(X))
n0=len(x[y<fx])
base=b-a
yest=n0/n*(base*altura)
print(yest)
```
