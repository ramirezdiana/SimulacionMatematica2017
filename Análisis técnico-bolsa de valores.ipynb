```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas_datareader import data
import datetime
```

```python
tickers=['GFINBURO.MX']
data_source='yahoo'
start_date=datetime.datetime(2011,4,18)
end_date=datetime.datetime(2017,11,14)
panel_data=data.DataReader(tickers,data_source,start_date,end_date)
panel_data
closes=panel_data.loc['Adj Close']
closes.plot(figsize=(14,6))
```
