```python
A=2
W=.5
B=.1
t=np.linspace(0,50,100)
plot.figure(figsize=(9,5))
plot.plot(t, A*np.cos(W*t)+B*np.sin(W*t),"--",lw=1,ms=4, label=r'$x(t)$')
plot.plot(t, W*B*np.cos(W*t)-W*A*np.sin(W*t),"ro-",lw=1,ms=4, label=r'$\dot{x(t)}$')
plot.xlabel('$t$',fontsize=20)
plot.legend(loc='best')
plot.show()
```

$$x(t)=A\cos(f*t)+B\sin(f*t)$$

mostrar en una gr√°fica oscilaciones con frecuencias a establecer en un arreglo f=frecuencia


```python
frecuencias=np.array([.1,.2,.5,.6])
plot.figure(figsize=(9,5))
for f in frecuencias:
    plot.plot(t,.5*np.cos(f*t)+.1*np.sin(f*t), '*-',lw=1,ms=4)
plot.xlabel('$t$',fontsize=14)
plot.ylabel('$x(t)$',fontsize=14)
plot.title('Oscilaciones con diferentes frecuencias', fontsize =16)
plot.show()
```

```python
from ipywidgets import *
from IPython.display import clear_output, display
```

```python
fig=plot.figure()
ax=fig.add_subplot(1,1,1)
def masa_resorte (t=0):
    ax.clear()
    ax.plot(.5*np.cos(.5*t)+.1*np.sin(.5*t),[0], 'ko', ms=10)
    ax.set_xlim(xmin=-.6,xmax=.6)
    ax.axvline(x=0,color='r')
    ax.axhline(y=0,color='grey',lw=1)
#   fig.canvas.draw()
    clear_output(wait=True)
    
    display(fig)
interact(masa_resorte,t=(0,50,.01));
plot.close()
```

```python
def theta_t(a,b,g,l,t):
    omega_0=np.sqrt(g/l)
    return a*np.cos(omega_0*t)+b*np.sin(omega_0*t)
```

```python
def pendulo_simple (t=0):
    fig=plot.figure(figsize=(5,5))
    ax=fig.add_subplot(1,1,1)
    x=2*np.sin(theta_t(.4,.6,9.8,2,t))
    y=-2*np.cos(theta_t(.4,.6,9.8,2,t))
    ax.plot(x,y,'ko', ms=10)
    ax.plot([0],[0],'rD')
    ax.plot([0,x], [0,y],'k-',lw=1)
    ax.set_xlim(xmin=-2.2,xmax=2.2)
    ax.set_ylim(ymin=-2.2,ymax=.2)
    fig.canvas.draw()
interact_manual(pendulo_simple,t=(0,10,.01));
```
